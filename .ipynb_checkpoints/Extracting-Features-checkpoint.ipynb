```python
import librosa
import matplotlib.pyplot as plt
```

## 1. Importing speech file and sampling

```python
y, sr = librosa.load("./salam.wav")
print("first 5 values of y: ", y[:5])
print("y shape: ", y.shape)
print("type of y: ", type(y))
print("sampling rate: ", sr)
```

Let's plot our signal

```python
plt.plot(y)
plt.title("Signal")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
```

## 2. Preprocessing

### A. Frame blocking

To frame a specific audio signal, we have to specify to values `N` and `M`. `N` is the length of each frame, and `M` is the length of hop. We usually specify `M < N` to have overlap between our frames.

First, let's calculate the duration of audio file. To do that, we have to know that `sampling rate` is the number of samples per second. So, if we divide all the samples by `sr`, it will give us the duration of audio file.

```python
audio_duration = y.shape[0] / sr
print("duraction of audio file: ", audio_duration, "s")
```

Let's set N equal to length of 30 ms, and M equal to length of 10 ms

```python
N = sr * 30 // 1000
M = sr * 10 // 1000
print("N: ", N, "samples")
print("M: ", M, "samples")
```

Now that we specify N and M, let's frame the audio file

```python
frames = librosa.util.frame(y, frame_length=N, hop_length=M, axis=0, writeable=True)
print("number of frames: ", len(frames))
print("shape of first frame: ", frames[0].shape)
```

Let's specify a frame to use it in extraction

```python
chosen_frame = 89
```

### B. Windowing

After framing the audio file, we have to multiply each frame to a window. Here, we will use ***hamming window***.

```python
window = librosa.filters.get_window('hamming', N)
plt.plot(window)
plt.title("Hamming window")
plt.ylabel("Amplitude")
plt.xlabel("Samples")
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1,2,1)
plt.plot(frames[chosen_frame])
plt.title(f"Frame {chosen_frame} before windowing")
plt.xlabel("Samples")
plt.ylabel("Amplitude")

for i in range(len(frames)):
    frames[i] = frames[i] * window
    
plt.subplot(1,2,2)
plt.plot(frames[chosen_frame])
plt.title(f"Frame {chosen_frame} after windowing")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
```

### C. Preemphasis

After windowing, we will pre emphasize on our frames.

preemphasis' formula is:
`S(n) = S(n) - alpha . S(n - 1)`

Note: the default value of alpha in `librosa.effects.preemphasis` is 0.97

```python
import numpy as np
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.plot(frames[chosen_frame])
plt.title(f"Frame {chosen_frame} before preemphesis")
plt.xlabel("Samples")
plt.ylabel("Amplitude")

preemphasis_frames = np.zeros(frames.shape)

for i in range(len(frames)):
    preemphasis_frames[i] = librosa.effects.preemphasis(frames[i])
    
plt.subplot(1,2,2)
plt.plot(preemphasis_frames[chosen_frame])
plt.title(f"Frame {chosen_frame} after preemphesis")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
```

## 3. Extract Features

### A. Time-domain features

#### a. Mean

Computing mean of a frame is not very hard. the formula is:

$ \mu = \frac{1}{N} \sum \limits _{n=0} ^{N-1} s(n) $

```python
import numpy as np
mean = np.sum(frames, axis=1) / N
print("shape of mean is ", mean.shape)
print(f"mean of frame {chosen_frame} is ", mean[chosen_frame])
```

#### b. Energy

The formula of computing energy is:

$ E = \frac{1}{N} \sum \limits _{n=0} ^{N-1} s(n) * s(n)$

```python
energy = np.sum(frames * frames, axis=1) / N
print("shape of energy is ", energy.shape)
print(f"energy of frame {chosen_frame} is ", energy[chosen_frame])
```

#### c. ZCR (Zero Crossing Rate)

Zero crossing rate can easily be computed by the following formula:

$ ZCR = \frac{1}{N} \sum \limits _{n=1} ^{N-1} \frac{|sgn[s(n)] - sgn[s(n-1)]|}{2} $

$ sgn[s(n)] = 1 $ if $ s(n) \geq 0 $

$ sgn[s(n)] = -1 $ if $ s(n) < 0 $

```python
ZCR = np.zeros(len(frames))

for i in range(len(frames)):
    for j in range(1, frames.shape[1]):
        ZCR[i] += ((abs(frames[i,j] - frames[i,j-1])) / 2)
    ZCR[i] /= N

print("The shape of ZCR is ", ZCR.shape)
print(f"Zero crossing rate of frame {chosen_frame} is ", ZCR[chosen_frame])
```

#### d. Auto-correlation

the formula of calculating auto correlation coefficients is:

$ r(\eta) = \frac{1}{N} \sum \limits _{n=0} ^{N-1} s(n) * s(n - \eta) $

```python
plt.figure(figsize=(12, 6))
plt.subplot(1,2,1)
plt.plot(frames[chosen_frame])
plt.title(f"Frame {chosen_frame}")
plt.xlabel("Samples")
plt.ylabel("Amplitude")

AC = np.zeros(frames.shape)

for i in range(len(frames)):
    AC[i,:] = librosa.autocorrelate(frames[i])

print("The shape of AC is ", AC.shape)

plt.subplot(1,2,2)
plt.plot(AC[chosen_frame])
plt.title(f"Auto-Correlation of Frame {chosen_frame}")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
```

Also, to improve the result of auto correlation of a signal, we can:
* Multiply the signal to a power of an odd number
* Center Clipping
* 3-level Center Clipping

##### Multiply the signal to a power of an odd number


```python
powered_frames = frames ** 3

plt.figure(figsize=(12, 6))
plt.subplot(1,2,1)
plt.plot(powered_frames[chosen_frame])
plt.title(f"Powered Frame {chosen_frame}")
plt.xlabel("Samples")
plt.ylabel("Amplitude")

PAC = np.zeros(frames.shape)

for i in range(len(powered_frames)):
    PAC[i,:] = librosa.autocorrelate(powered_frames[i])

print("The shape of PAC is ", PAC.shape)

plt.subplot(1,2,2)
plt.plot(PAC[100])
plt.title(f"Auto-Correlation of Powered Frame {chosen_frame}")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
```

##### Center Clipping

```python
c = 0.02
cc_frames = (np.abs(frames) < c)*0 + (frames >= c)*(frames-c) + (frames <= -c)*(frames+c)

plt.figure(figsize=(12, 6))
plt.subplot(1,2,1)
plt.plot(cc_frames[chosen_frame])
plt.title(f"Center Clipped Frame {chosen_frame}")
plt.xlabel("Samples")
plt.ylabel("Amplitude")

CCAC = np.zeros(frames.shape)

for i in range(len(cc_frames)):
    CCAC[i,:] = librosa.autocorrelate(cc_frames[i])

print("The shape of CCAC is ", CCAC.shape)

plt.subplot(1,2,2)
plt.plot(CCAC[chosen_frame])
plt.title(f"Auto-Correlation of Center Clipped Frame {chosen_frame}")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
```

##### 3 Level Center Clipping

```python
c = 0.02
tlcc_frames = (np.abs(frames) < c)*0 + (frames >= c)*1 + (frames <= -c)*1

plt.figure(figsize=(12, 6))
plt.subplot(1,2,1)
plt.plot(tlcc_frames[chosen_frame])
plt.title(f"3 Level Center Clipped Frame {chosen_frame}")
plt.xlabel("Samples")
plt.ylabel("Amplitude")

TLCCAC = np.zeros(tlcc_frames.shape)

for i in range(len(tlcc_frames)):
    TLCCAC[i,:] = librosa.autocorrelate(tlcc_frames[i])

print("The shape of TLCCAC is ", TLCCAC.shape)

plt.subplot(1,2,2)
plt.plot(CCAC[chosen_frame])
plt.title(f"Auto-Correlation of 3 Level Center Clipped Frame {chosen_frame}")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
```

#### e. AMDF (Average Magnitude Difference Function)

The formula of computing AMDF is:

$ AMDF(\eta) = \frac{1}{N} \sum \limits _{n=0} ^{N-1} |s(n) - s(n - \eta)| $ where $ 0 \leq \eta \leq N-1 $

```python
import copy
AMDF = np.zeros(frames.shape)
print(frames.shape, N)

for i in range(len(frames)):
    window = copy.deepcopy(frames[i])

    for eta in range(N):
        AMDF[i, eta] = np.sum(np.abs(frames[i] - window)) / N
        window = np.insert(window[:N-1], 0, 0)
        assert window.shape[0] == N
        assert window[0] == 0
        assert AMDF[i, eta].shape == ()
    
print("Shape of AMDF is ", AMDF.shape)
        
plt.figure(figsize=(12, 6))
plt.subplot(1,2,1)
plt.plot(frames[chosen_frame])
plt.title(f"Frame {chosen_frame}")
plt.xlabel("Samples")
plt.ylabel("Amplitude")

plt.subplot(1,2,2)
plt.plot(AMDF[chosen_frame])
plt.title(f"AMDF of Frame {chosen_frame}")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
```

### B. Frequency-domain features

#### a. FFT (Fast Fourier Transform)

To calculate FFT, we will use `fft()` function from `scifi.fft`.

The input will be the vector that we want to calculate the fft from.

```python
from scipy.fft import fft
fft_values = fft(frames[chosen_frame])
abs_of_fft_values = np.abs(fft_values)
log_of_abs_of_fft_values = np.log(abs_of_fft_values)

plt.figure(figsize=(12, 6))

plt.subplot(1,2,1)
plt.plot(frames[chosen_frame])
plt.title(f"Frame {chosen_frame}")
plt.xlabel("Samples")
plt.ylabel("Amplitude")

plt.subplot(1,2,2)
plt.plot(log_of_abs_of_fft_values)
plt.title(f"FFT of frame {chosen_frame}")
plt.xlabel("Frequency[Hz]")
plt.ylabel("Magnitude [dB]")
```

#### b. LPC (Linear predictive coding)

the formula is:

$ s(n) = \sum \limits _{k=1} ^{P} a _{k} s(n - k) $

but for simplicity, we will use `librosa.lpc` function, and we will pass the original signal to it, and then, we will plot it. In the following block, `order` is equivalent to `P` in the formula.

```python
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.plot(y[chosen_frame * M: chosen_frame * M + N])
plt.title(f"Original signal of frame {chosen_frame}")

plt.subplot(1,2,2)
lpc_co = librosa.lpc(y[chosen_frame * M: chosen_frame * M + N], order=32)
plt.plot(lpc_co)
plt.title(f"LPC coefficients of frame {chosen_frame}")
```

#### c. Cepstral Analysis

Consider our signal is equal to `s(n)`. To calculate the cepstrum coefficients, we have to calculate:
$ C _{s}(n) = FFT^{-1}[Log|FFT(s(n))|] $

So, we have to apply a `fft` on `s(n)`, then get the `absolute` of that, and after that, get the `Log`, and at the end, get `FFT inverse` from the result.

```python
from scipy.fft import fft, ifft

plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.plot(frames[chosen_frame])
plt.title(f"Original signal of frame {chosen_frame}")

S_k = fft(frames[chosen_frame])
absolute_values = np.abs(S_k)
log_of_values = np.log(absolute_values)
C_s = ifft(log_of_values)
plt.subplot(1,2,2)
plt.plot(C_s)
plt.title(f"cepstrum coefficients of frame {chosen_frame}")
```

#### d. Spectrum Estimation

##### i. Spectrum Estimation based on FFT

```python
S_k = fft(preemphasis_frames[chosen_frame])
abs_of_S_k = np.abs(S_k)
log_of_S_k = np.log(abs_of_S_k)

plt.figure(figsize=(12, 6))
plt.subplot(1,2,1)
plt.plot(preemphasis_frames[chosen_frame])
plt.title(f"Preemphasised frame {chosen_frame}")
plt.xlabel("Samples")
plt.ylabel("Amplitude")

plt.subplot(1,2,2)
plt.plot(log_of_S_k[:(len(log_of_S_k) // 2)])
plt.title(f"FFT coefficients of frame {chosen_frame}")
plt.xlabel("Frequency [Hz]")
plt.ylabel("Magnitude [dB]")
```

* The `pitch frequency` is the distance between the huge ripples of FFT Coefficients
* The formants are the maximum points of the push of FFT Coefficients, which are easily distinguishable

##### ii. Spectrum Estimation based on LPC

```python
lpc_co = librosa.lpc(y[chosen_frame * M: chosen_frame * M + N], order=120)
S_k = fft(lpc_co)
abs_of_S_k = np.abs(S_k)
reversed_S_k = np.flip(abs_of_S_k[:len(abs_of_S_k)//2])
print(abs_of_S_k[:10])
print(reversed_S_k[:10])
# log_of_S_k = np.log(abs_of_S_k)

plt.figure(figsize=(12, 6))
plt.subplot(1,2,1)
plt.plot(preemphasis_frames[chosen_frame])
plt.title(f"Preemphasised frame {chosen_frame}")
plt.xlabel("Samples")
plt.ylabel("Amplitude")

plt.subplot(1,2,2)
plt.plot(reversed_S_k)
plt.title(f"LPC Spectrum of frame {chosen_frame}")
plt.xlabel("Frequency [Hz]")
plt.ylabel("Magnitude [dB]")
```

#### e. Pitch Estimation

##### i. Pitch Estimation based on Auto Correlation

```python
max_first_10ms = max(frames[chosen_frame][:N//3])
max_last_10ms  = max(frames[chosen_frame][2*N//3:])

K = 0.68
C = K * min(max_first_10ms, max_last_10ms)

center_clipped_frame = np.zeros(N)
center_clipped_frame += ((frames[chosen_frame] > C) * (frames[chosen_frame] - C))
center_clipped_frame += ((frames[chosen_frame] < -C) * (frames[chosen_frame] + C))

acc = librosa.autocorrelate(center_clipped_frame)


plt.figure(figsize=(12, 12))

plt.subplot(2,2,1)
plt.plot(frames[chosen_frame])
plt.title(f"Original Signal of frame {chosen_frame}")

plt.subplot(2,2,2)
plt.title("Signal after center clipping")
plt.plot(center_clipped_frame)

plt.subplot(2,2,3)
plt.plot(acc)
plt.title("Auto correlation")
```

Now Let's find $I _{pos}$

$I _{pos}$ will be the distance to the first peak

```python
from matplotlib import lines
acc_cp = np.copy(acc)

# Remove first part of auto correlation
acc_cp[:10] = 0

# Find maximum
maximum = max(acc_cp)

# Apply a center clipping
acc_cp = (np.abs(acc_cp) > maximum * 0.68)

# Find first peak
I_pos = 0
for i in range(len(acc_cp)):
    if acc_cp[i] != 0:
        I_pos = i
        break

plt.plot(acc)
plt.plot([I_pos, I_pos], [min(acc), max(acc)], color="orange")
vertical_line = lines.Line2D([], [], color='orange', marker='|', linestyle='None',
                          markersize=10, markeredgewidth=1.5, label='I pos')

plt.legend(handles = [vertical_line])
```

After finding $I _{pos}$, the last step is to calculate $ F _{pitch} $ considering the formula bellow:

$F _{pitch} = \frac{F _{s}}{I _{pos}}$

```python
F_pitch = sr / I_pos

print(f"Frequency of pitch is equal to '{F_pitch} Hz'")
```
